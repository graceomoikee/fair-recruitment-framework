
# Baseline Pipeline Smoke Test

This notebook validates the baseline data loading and preprocessing pipeline.

It ensures:
- Dataset is correctly loaded
- Feature, label, and protected attribute splits are valid
- Class and group distributions are as expected

This serves as a sanity check before running full experiments.

```python
import sys
from pathlib import Path

# Add project root to Python path
project_root = Path.cwd().parent
sys.path.append(str(project_root))


```

```python
from src.data.loader import load_hr_analytics

data_path = Path(project_root) / "data" / "raw" / "HR_Analytics.csv"
ds = load_hr_analytics(str(data_path))

```

```python
# Verify dataset structure
print("Feature matrix shape:", ds.X.shape)
print("Target shape:", ds.y.shape)
print("Protected attribute shape:", ds.A.shape)
```

```python
# Class distribution (Attrition)
print("Target counts:\n", ds.y.value_counts(dropna=False))
print("\nTarget proportions:\n", ds.y.value_counts(normalize=True))
```

```python
# Protected attribute distribution (Gender)
print("Protected attribute counts:\n", ds.A.value_counts(dropna=False))
print("\nProtected attribute proportions:\n", ds.A.value_counts(normalize=True))
```

```python
import pandas as pd

# Analyse outcome distribution across groups
tmp = pd.DataFrame({"y": ds.y, "A": ds.A})
group_stats = tmp.groupby("A")["y"].agg(["mean", "count"])

print("Group-wise outcome statistics:\n", group_stats)
```
